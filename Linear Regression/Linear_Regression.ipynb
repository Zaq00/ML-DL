```python
import pandas as pd

data = pd.read_csv('train.csv')
```

```python
# import matplotlib.pyplot as plt
null_counts = data.isnull().sum()
total_counts = len(data)
null_percentages = (null_counts / total_counts) * 100
non_null_percentages = 100 - null_percentages

# # Create separate pie plots for each column
# for column in data.columns:
#     fig, ax = plt.subplots()
#     ax.pie([null_percentages[column], non_null_percentages[column]],
#            labels=['Null', 'Non-Null'], autopct='%1.1f%%', startangle=90)
#     ax.set_title(f"{column} - Null vs. Non-Null")
#     plt.show()
```

```python
for i in data.columns:
    if null_percentages[i] >= 70:
        data.drop(i, axis=1, inplace=True)
```

```python
data
```

Replacing NaN Values

Mode for type Object

Mean for Numerical Values

```python
for i in data.columns:
    if data[i].dtype == 'O':
        data[i] = data[i].fillna(data[i].mode()[0])
    else:
        data[i] = data[i].fillna(data[i].mean())
```

```python
data.isnull().sum().to_dict()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'O':
        data[i] = encoder.fit_transform(data[i])
```

```python
data
```

```python
data.drop('Id', axis=1, inplace=True)
```

```python
X = data.drop('SalePrice', axis=1)
Y = data['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X, Y)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(Y, prediction)
rmse = mean_squared_error(Y, prediction, squared=False)
mae = mean_absolute_error(Y, prediction)
r2 = r2_score(Y, prediction)
```
